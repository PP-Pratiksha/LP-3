```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sales_data_sample.csv', encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
corr=df.corr()
sns.heatmap(corr,annot=True)
```

```python
df= df[['PRICEEACH', 'MSRP']]
```

```python
df
```

```python
df.isna().any()
```

```python
from sklearn.cluster import KMeans
inertia=[KMeans(n_clusters=i,init='k-means++',random_state=42).fit(df).inertia_ for i in range(1,11)]
plt.figure(figsize=(12,10))
plt.plot(range(1,11),inertia)
plt.xlabel("No of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 42)
y_kmeans = kmeans.fit_predict(df)
y_kmeans
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x=df['PRICEEACH'], y=df['MSRP'], hue=y_kmeans)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c = 'red', label = 'Centroids', marker='*')
plt.legend()
```
